```python
df=data[['Age','LanguageWorkedWith','ConvertedComp']]
```

```python
df.head()
```

```python
df.sort_values('Age',ascending=False)
```

```python
df.dropna(inplace=True)
df['Age']=df['Age'].astype(int)
```

```python
df.head(20)
```

```python
df.sort_values('Age',ascending=False)
```

```python
mask=(df['Age']>=18) & (df['Age']<=55)
```

```python
df_new=df[mask]
```

```python
df_new.sort_values('Age')
```

```python
df_new['python']=df_new['LanguageWorkedWith'].str.contains('Python')
```

```python
filt=df_new['python']==True
df_new_python=df_new[filt]
```

```python
df_new_python.rename(columns={'ConvertedComp':'Pythoncomp'},inplace=True)
df_new_python.drop(columns=['LanguageWorkedWith','python'],inplace=True)
```

```python
df_new['javascript']=df_new['LanguageWorkedWith'].str.contains('JavaScript')
```

```python
filt=df_new['javascript']==True
df_new_js=df_new[filt]
```

```python
df_new_js
```

```python
df_new_js.rename(columns={'ConvertedComp':'JScomp'},inplace=True)
df_new_js.drop(columns=['LanguageWorkedWith','python','javascript'],inplace=True)
```

```python
df_new_python
```

```python
df_new_js
```

```python
df_new.rename(columns={'LanguageWorkedWith':'All'},inplace=True)
```

```python
df_new
```

```python
df_new.drop(columns=['All','javascript','python'],inplace=True)
```

```python
df_new
```

```python
fin=pd.concat([df_new_python,df_new_js])
```

```python
final=pd.concat([fin,df_new])
```

```python
final
```

```python
data=pd.pivot_table(final,index='Age',values=['Pythoncomp','JScomp','ConvertedComp'],aggfunc='median').reset_index()
```

```python
data
```

```python
all_median=data['ConvertedComp'].median()
```

```python
age=data['Age']
all_salary=data['ConvertedComp']
py_salary=data['Pythoncomp']
js_salary=data['JScomp']
```

```python
plt.plot(age,all_salary,'--r',label='All')
plt.plot(age,py_salary,label='Python')
plt.fill_between(age,py_salary,all_median,
                 where=(py_salary>all_median),interpolate=True,alpha=0.25)
plt.fill_between(age,py_salary,all_median,
                 where=(py_salary<=all_median),color='red',interpolate=True,alpha=0.25)
plt.title('Median salary by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.show()
```
